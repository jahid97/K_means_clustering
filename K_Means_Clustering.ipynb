```python
#Clustering with K-Means
#---------------------------------------------------#

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
# X= -2 * np.random.rand(100,2)
# X1 = 1 + 2 * np.random.rand(50,2)
# X[50:100, :] = X1

X = np.array([((7,5)),((5,7)),((7,7)),((3,3)),((4,6)),((1,4)),((0,0)),((2,2)),((8,7)),((6,8)),((5,5)),((3,7))])

plt.scatter(X[ : , 0], X[ :, 1], s = 50, c = 'b')
plt.show()


```

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=2)
Kmean.fit(X)
```

```python
Kmean.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c='b')
plt.scatter(1.5  , 2.25 , s=150, c='g', marker='s')
plt.scatter( 5.625        , 6.5 , s=50, c='r', marker='o')

plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.spatial.distance import cdist
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

iris = load_iris()


res = list()
n_cluster = range(1,10)
for n in n_cluster:
    kmeans = KMeans(n_clusters=n)
    kmeans.fit(X)
    res.append(np.average(np.min(cdist(X, kmeans.cluster_centers_, 'euclidean'), axis=1)))

plt.plot(n_cluster, res)
plt.title('elbow curve')
plt.show()
```
